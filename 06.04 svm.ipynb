# Q1. What is the mathematical formula for a linear SVM?

 A linear Support Vector Machine (SVM) is a type of machine learning algorithm that is commonly used for classification tasks. The mathematical formula for a linear SVM can be written as:

         y = w^T x + b

where:

y is the predicted class label for the input sample x
w is the weight vector
b is the bias term
The goal of the linear SVM is to find the weight vector w and the bias term b that will maximize the margin between the decision boundary and the closest data points from both classes. The decision boundary is the hyperplane that separates the data points of different classes.

To train a linear SVM, we use an optimization algorithm to minimize the objective function:

minimize (1/2) ||w||^2
subject to y_i(w^T x_i + b) >= 1 for all i
where:

||w|| is the Euclidean norm of the weight vector
y_i is the class label for the i-th training sample
x_i is the i-th training sample


# Q2. What is the objective function of a linear SVM?

The objective function of a linear Support Vector Machine (SVM) is used to find the optimal hyperplane that separates the data points of different classes while maximizing the margin between them. The objective function is formulated as an optimization problem that the SVM seeks to minimize. It involves two main components: the margin maximization term and the classification error term.

Given a labeled dataset with input features X and corresponding binary class labels y (where y can be either -1 or 1 for binary classification), the objective function of a linear SVM is represented as follows:

minimize ||w||^2 + C * Σ(max(0, 1 - y_i * (w^T * x_i + b)))
   w,b


where:

||w||^2: The L2-norm (Euclidean norm) of the weight vector w. This term represents the margin maximization objective, as it encourages the SVM to find a hyperplane that maximizes the distance between the two classes' closest data points, known as the margin.
C: The hyperparameter that controls the trade-off between maximizing the margin and minimizing the classification error. It acts as a regularization parameter, preventing overfitting. A larger C value penalizes misclassifications more severely, leading to a narrower margin, while a smaller C value allows more misclassifications but results in a wider margin.
x_i: The feature vector of the i-th data point.
y_i: The corresponding class label of the i-th data point (either -1 or 1).
w: The weight vector that represents the normal vector to the hyperplane.
b: The bias term or the intercept of the hyperplane.
The first term ||w||^2 aims to maximize the margin by controlling the magnitude of the weight vector w. The larger the magnitude of w, the narrower the margin, and vice versa.

The second term Σ(max(0, 1 - y_i * (w^T * x_i + b))) represents the classification error term. This term enforces that the data points are classified correctly, meaning they are on the correct side of the hyperplane with a margin of at least 1. If a data point is correctly classified (i.e., y_i * (w^T * x_i + b) >= 1), the term contributes 0 to the objective function. However, if a data point is misclassified (i.e., y_i * (w^T * x_i + b) < 1), the term adds a positive value to the objective function, penalizing the misclassification.

The SVM aims to minimize this objective function by adjusting the values of w and b, and the optimal values of w and b correspond to the hyperplane that best separates the data points with the widest margin.

# Q3. What is the kernel trick in SVM?

The kernel trick is a technique used in Support Vector Machines (SVMs) to handle non-linearly separable data by implicitly transforming the data into a higher-dimensional feature space. It allows SVMs to find a separating hyperplane in this higher-dimensional space, even if the data points are not linearly separable in the original feature space.

In simple terms, the kernel trick enables SVMs to effectively deal with complex data distributions and non-linear decision boundaries without explicitly computing the coordinates of data points in the higher-dimensional space.

The kernel trick works as follows:

Original Feature Space: In the original feature space, the data points may not be linearly separable. 

Mapping to Higher-Dimensional Space: To make the data points separable, the kernel trick applies a mathematical function (called the kernel function) to implicitly transform the data points into a higher-dimensional space. In this higher-dimensional space, the data points can become separable by a hyperplane.

Finding the Hyperplane: The SVM algorithm works with the transformed data in the higher-dimensional space to find the optimal hyperplane that best separates the data points of different classes while maximizing the margin.

The choice of the kernel function is crucial in the kernel trick. Commonly used kernel functions include:

a. Linear Kernel: The linear kernel (K(x, x')) is the default kernel and performs a standard dot product between data points. It is suitable for linearly separable data.

b. Polynomial Kernel: The polynomial kernel (K(x, x') = (1 + x^T * x')^d) transforms data into a higher-dimensional space using polynomial terms. The 'd' parameter controls the degree of the polynomial.

c. Radial Basis Function (RBF) Kernel: The RBF kernel (K(x, x') = exp(-gamma * ||x - x'||^2)) transforms data using a Gaussian-like function. The 'gamma' parameter controls the smoothness of the decision boundary.

d. Sigmoid Kernel: The sigmoid kernel (K(x, x') = tanh(alpha * x^T * x' + c)) transforms data using the hyperbolic tangent function.

By using different kernel functions, SVMs can handle various types of data distributions and create non-linear decision boundaries, making the SVM algorithm more versatile and powerful for a wide range of machine learning tasks. The kernel trick is one of the key reasons why SVMs are effective in handling complex data.

# Q4. What is the role of support vectors in SVM Explain with example

In Support Vector Machines (SVM), support vectors play a crucial role in defining the decision boundary (hyperplane) that separates different classes of data points. Support vectors are the data points from the training dataset that are closest to the decision boundary and have the most influence on its position and orientation.

When training an SVM, the main objective is to find the optimal hyperplane that maximizes the margin between the support vectors of different classes while minimizing the classification error. The margin is the perpendicular distance between the decision boundary and the closest support vectors from each class.

When we apply an SVM to this data, it tries to find the best-fitted decision boundary (hyperplane) that separates the two classes. The hyperplane is positioned in such a way that it maximizes the margin between the support vectors of the two classes.

The support vectors are the data points that are closest to the decision boundary on either side.

The position and orientation of the decision boundary are determined by the support vectors. The SVM ensures that the support vectors from each class lie on the correct side of the decision boundary with a margin of at least 1.

The non-support vector data points (the other points) do not affect the position of the decision boundary, and their distances from the boundary are not as important. The SVM only considers the support vectors in defining the decision boundary, making it computationally efficient, especially for large datasets.

In summary, support vectors are the critical data points that influence the decision boundary's position and help define the maximum margin between classes. They are essential in SVM because they represent the most challenging and informative data points that guide the training process to find the optimal hyperplane for classification tasks.

# Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?

Hyperplane:
In Support Vector Machines (SVM), a hyperplane is a decision boundary that separates data points of different classes in a feature space. In 2D, a hyperplane is a straight line that separates two classes, while in 3D, it becomes a flat plane. In higher dimensions, a hyperplane is a hyper-surface. The objective of SVM is to find the optimal hyperplane that maximizes the margin between the classes.

Marginal Plane:
The marginal plane is a parallel plane to the optimal hyperplane, but it is located exactly on the closest support vectors from each class. It plays a crucial role in determining the margin width, which is the distance between the optimal hyperplane and the marginal plane.

Soft Margin:
In SVM, a soft margin allows for some misclassification of data points to achieve a wider margin between classes. This is useful when the data is not perfectly separable. The soft margin introduces a slack variable that permits a certain degree of misclassification while still aiming to maximize the margin.

Hard Margin:
Contrary to the soft margin, a hard margin in SVM demands that all data points be correctly classified without any misclassification. This approach works best when the data is perfectly separable by a hyperplane with no overlap between classes.

In summary, the hyperplane is the decision boundary that separates data points, the marginal plane determines the margin width, the soft margin allows for some misclassification to achieve better separation, and the hard margin requires no misclassification for perfect separation. The choice of margin type depends on the data characteristics and the task's requirements.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = datasets.load_iris()

# making dependent and independent variable
X = iris.data[:, :2]  
y = iris.target
```

```python
# Split the dataset into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train a linear SVM classifier on the training set
C = 1.0  # Regularization parameter
svm_classifier = SVC(kernel='linear', C=C)


svm_classifier.fit(X_train, y_train)
```

```python
# Predict the labels for the testing set
y_pred = svm_classifier.predict(X_test)

# Compute the accuracy of the model on the testing set
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy",accuracy)
```

```python
# Plot the decision boundaries of the trained model using two features
# To plot the decision boundaries, we will create a meshgrid of points in the feature space
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
```

```python
# Predict the class for each point in the meshgrid
Z = svm_classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
# Plot the contour plot and scatter plot for the two features
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k', s=100)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundaries of Linear SVM')
plt.show()
```
